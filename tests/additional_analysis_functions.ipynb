```python
from astrosceni.image import Image
from astrosceni.stars_filter import StarsFilter
from astrosceni.mufinder import MuFinder
```

```python
nb = Image("data/bs_h_ave_wcs.fits")
bb = Image("data/bs_r_ave_wcs.fits")

nb.cropPixels(100, -100, 100, -100)
bb.cropPixels(100, -100, 100, -100)

finder = MuFinder(nb, bb, mu_resolution=0.01, mu_range=(0, 3))
```

```python
# Prints the mu in the mu_linspace which is closest to skewness = 0
print(finder.getCorrespondingMu(skewness = 0))
```

```python
# Lines show the optimal mu and how its lined up with skewness = 0
finder.plotSkewnessVals()
```
